```python
x = tf.constant([[1], [2], [3], [4]], dtype=tf.float32)
y_true = tf.constant([[0], [-1], [-2], [-3]], dtype=tf.float32)
print(x)
print(y_true)
```

```python
linear_model = tf.layers.Dense(units=1)

y_pred = linear_model(x)
```

```python
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)

print(sess.run(y_pred))
```

```python
loss = tf.losses.mean_squared_error(labels=y_true, predictions=y_pred)
```

```python
print(sess.run(loss))
```

```python
optimizer = tf.train.GradientDescentOptimizer(0.01)
train = optimizer.minimize(loss)
```

```python
for i in range(1000):
  _, loss_value = sess.run((train, loss))
```

```python
def draw_text(np_img, message):
    image = Image.fromarray(np_img)
    draw = ImageDraw.Draw(image)
    (x, y) = (5, 5)
    color = 'rgb(255, 0, 0)' # black color
    draw.text((x, y), message, fill=color)
    return image
```

```python
dataset_tools = import_module('tools.material')

train_images, test_images, train_labels,  test_labels = dataset_tools.get_data()

# label_map = {
#     1: 'Cardboard',
#     2: 'Pamphlet',
#     3: 'Empty',
#     4: 'Plastic Foil',
#     5: 'Shredded Paper'
# }

# The right label of X.npy
label_map = {
    0: 'Cardboard',
    1: 'Empty',
    2: 'Shredded Paper',
    3: 'Pamphlet',
    4: 'Plastic Foil'
}

print(train_images.shape)
```

```python
for i in range(80, 90):
    img = train_images[i]
    label = train_labels[i]
    img = draw_text(img,  str(label) + ": " + label_map[label] )
    plt.figure(figsize=(5, 5))
    plt.imshow(img)
```
